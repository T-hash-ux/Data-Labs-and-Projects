## The Data



```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

file_path = "https://github.com/learn-co-curriculum/da-phase1-project-enterprise/raw/main/data/ames.csv"
df = pd.read_csv(file_path, index_col=0)
df.info()
```

```python
df.describe()
```

## Distribution of `SalePrice`

```python
df['SalePrice']
```

```python
df.describe()
```

```python
fig, ax = plt.subplots()
plt.hist(df['SalePrice'], bins=50)
ax.set_xlabel('Sales Price');
ax.set_ylabel('Number Of Houses')
ax.set_title('Ames, Iowa Housing Sales Between 2006 and 2010');
plt.vlines(df['SalePrice'].mean(),0, 80, label=f'Average SalePrice: {df["SalePrice"].mean():.2f}', color='red')
plt.vlines(df['SalePrice'].median(),0, 80, label=f"Median SalePrice: {df['SalePrice'].median():.2f}", color='orange')
plt.legend()
plt.show()
```

We can see from the Visualization above that there is a right skew that has a median of 163,000. The average sales price of a home in Iowa based on the data displayed is 180,921.20. We can see that the visualization displays outliers with the lowest being 34,900 sales price and the highest being 755,000 sales price.

## Differences between Subsets

In this section I will be creating a visialization to compare the SalePrice distributions based on the overall conditions of the homes. I will split the data into three conditions: below_average_condition, average_condition, and above_average_condition to determine the Price between the subsets.

```python
cond = df.groupby(['OverallCond', 'SalePrice']).sum() 
cond
```

```python
below_average_condition = df.loc[(df['OverallCond'] < 5), 'SalePrice']
average_condition = df.loc[(df['OverallCond'] == 5), 'SalePrice']
above_average_condition = df.loc[(df['OverallCond'] > 5), 'SalePrice']

fig, ax = plt.subplots()
ax.hist(above_average_condition,alpha=0.5,label='above average', bins=50)
ax.hist(average_condition,alpha=0.5,label='average condition', bins=50)
ax.hist(below_average_condition,alpha=0.5,label='below average', bins=50)

ax.set_xlabel('Sales Price');
ax.set_ylabel('Number Of Houses');
ax.set_title('Distribution of Sale Price Grouped by Condition');
ax.legend()
```

What we can interpret from the visualization is that houses that have a below average condition (< 5) have lower sale prices. The houses that have average to above conditions are priced significantly higher. Although the average homes are priced very high the above average condition homes are priced the highest. 

## Correlation between YearBuilt and `SalePrice`

```python
column1 = df['YearBuilt']
column2 = df['SalePrice']
correlation = column1.corr(column2)
correlation
```

```python
fig, ax = plt.subplots()
plt.scatter(df['YearBuilt'], df['SalePrice'])
ax.set_xlabel('YearBuilt');
ax.set_ylabel('SalePrice');
ax.set_title('Correlation between YearBuilt and SalePrice');
```

In the display above we can see that there ia a strong Curvilinear correlation between the year built and the Sale price. We know this because the direction of the two variable relationship is directed in a positive even though they overlap.

## Engineered Feature: Age

```python
df['Age'] = df['YrSold'] - df['YearBuilt']
df.describe()
```

```python
fig, ax = plt.subplots()
plt.scatter(df['SalePrice'], df['Age'])
ax.set_xlabel('SalePrice');
ax.set_ylabel('Age');
ax.set_title('Age vs SalePrice');
```

In the scatter plot above we see that there is a negative correlation. This impacts the Sales Price because as the age of the homes increase the Price of the homes decrease.
